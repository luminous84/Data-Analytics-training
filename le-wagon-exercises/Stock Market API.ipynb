```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
```

```python
import requests

url = "https://cloud.iexapis.com/stable/stock/aapl/chart/3m?token=pk_86a816e248cd4eb7a72b46a0823d21c5"
api_data = requests.get(url).json()
```

```python
apple_stock_df = pd.DataFrame.from_dict(api_data)
```

```python
print (apple_stock_df)
```

```python
apple_stock_df['date'] = pd.to_datetime(apple_stock_df['date'], format="%Y-%m-%d")
```

```python
apple_stock_df = apple_stock_df.set_index('date')
```

```python
print (apple_stock_df)
```
