## Yearly Gas production

How much gas has been produced yearly globally and in every state of the US?

```python
yearly_gas_df = gas_df.groupby(gas_df['Month'].dt.year).sum()
yearly_gas_df
```

Highest Gas Production by Year in U.S. in Million

```python
yearly_gas_df.sort_values("U.S.", ascending=False).head(10).filter(items=['U.S.']) / 1_000_000
```

## Plotting some data

```python
plot = yearly_gas_df.filter(items=['U.S.']).plot(kind="bar")
plot.set_xlabel("Year")
plot.set_ylabel("Gas Production")
```

Only use years with complete data (discarding 2008 and 2018)

```python
full_years = np.logical_and(yearly_gas_df.index >= 2009, yearly_gas_df.index <= 2017)
full_years
```

```python
full_years_gas_df = yearly_gas_df [full_years]
full_years_gas_df
```

```python
plot = full_years_gas_df.filter(items=['U.S.']).plot(kind="bar")
plot.set_xlabel("Year")
plot.set_ylabel("Gas Production")
```

## Comparing with Crude Oil Production

```python
file = "U.S._crude_oil_production.csv"
oil_df = pd.read_csv(file, decimal=",")
```

```python
oil_df.head(5)
```

Month column into a datetime object:

```python
oil_df['Month'] = pd.to_datetime(oil_df['Month'], format='%Y-%m-%d')
```

```python
oil_df.dtypes[0:2]
```

Then create two dataframes yearly_gas and yearly_oil by grouping by year, summing and filtering the global production for both commodities. Rename the columns to Gas and Crude oil.

```python
yearly_gas = gas_df.groupby(gas_df['Month'].dt.year).sum().filter(items=['U.S.'])
yearly_gas.columns = [ 'Gas' ]
yearly_gas.head(5)
```

```python
yearly_oil = oil_df.groupby(oil_df['Month'].dt.year).sum().filter(items=['U.S. Crude Oil '])
yearly_oil.columns = [ 'Crude Oil' ]
yearly_oil.head(5)
```

```python
yearly_merged = pd.concat([yearly_gas, yearly_oil], axis = 'columns')
yearly_merged
```

```python
plot = yearly_merged.plot(kind ='bar')
plot.set_xlabel("Year")
plot.legend(['Gas (Millions of Cubic feet)', 'Crude Oil (Thousands of barrels)'])
```
